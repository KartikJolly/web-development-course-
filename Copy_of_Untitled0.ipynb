## ZOMATO DATASET EXPLORATORY DATA ANALYSIS


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
sales = pd.read_csv('diwalisales.csv', encoding='latin-1')
```

```python
sales
```

```python
sales.shape
```

```python
sales.info()
```

```python
pd.isnull(sales).sum()
```

```python
sales.dropna(inplace = True)
```

```python
pd.isnull(sales).sum()
```

```python
sales.describe()
```

```python
sales[['Age','Amount', 'Orders']].describe()
```

```python
# EXPLORATORY DATA ANALYSIS
```

```python
c = sns.countplot(x = 'Gender', data = sales)
for bars in c.containers:
    c.bar_label(bars)
```

```python
sales.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)
```

```python
sns.countplot(x = 'Age Group', data = sales, hue = 'State')
```

```python
sales_state = sales.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.barplot(x = 'State', y = 'Amount', data = sales_state)
```

```python
a = sns.countplot(x = 'Marital_Status', data = sales)
for bars in a.containers:
    a.bar_label(bars)
```

```python
a = sns.countplot(x = 'Occupation', data = sales)
for bars in a.containers:
    a.bar_label(bars)
```

```python
a = sns.countplot(x = 'Product_Category', data = sales)
sns.set(rc = {'figure.figsize':(20,2)})
for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_state = sales.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.barplot(x = 'Product_Category', y = 'Amount', data = sales_state)
```

```python
sales.groupby('Product_ID')['Orders'].sum().sort_values(ascending = False).head(10)
```
